```python
import pandas as pd
import genanki as ga
import os
```

```python
core = pd.read_csv("core_wip_jun29.csv").drop(["Unnamed: 0","Vocabulary-Kanji","Vocabulary-Kana",
                                               "Vocabulary-English","pitchaccent","Sentence-Image"],axis=1)
```

```python
core['pitch-video'] = core["Sentence-Audio"].str.replace(".mp3",".mp4")
```

```python
videos = ["[sound:"+i+"]" for i in os.listdir("final")]
```

```python
core
```

```python
from random import randrange
def rand_id():
    return randrange(1<<30,1<<31)
```

```python
backstr = \
"""
{{Expression}}

<hr id=answer>

{{Pitch Video}}<br>
{{Pronunciation}}<br>


<a onclick="replay()">click to replay</p>

<script>
	 function replay() {
       var elem = document.querySelector(".soundLink, .replaybutton");
       if (elem) { elem.click(); }
   }
   
</script>

"""
pitchmodel = ga.Model(rand_id(),
                 'Sentence pitch model',
                fields=[{'name':'Expression'},
                        {'name':'Translation'},
                        {'name':'Audio'},
                        {'name':'Pronunciation'},
                        {'name':'Sentence Pitch Drops'},
                        {'name':'Pitch Video'},
                        {'name':'Pitch Video (HTML5)'}
                        
                ],
                templates=[
                    {
                        'name':'Sentence pitch production',
                        'qfmt':'Expression',
                        'afmt':backstr
                    }
                ])
```

```python
deck = ga.Deck(rand_id(),'Core2k Sentence Pitch')
```

```python
for i in range(2000):
    data = list(core.loc[i])
    data[3] = ""
    soundname = data[-1]
    data.append(soundname[soundname.index(":")+1:-1])
    note = ga.Note(model=pitchmodel,fields=data)
    deck.add_note(note)
```

```python
audio_files = ["audios"+"/"+i for i in os.listdir("audios")]
video_files = ["final"+"/"+i for i in os.listdir("final")]
```

```python
pkg = ga.Package(deck)
```

```python
pkg.media_files = audio_files+video_files
```

```python
pkg.write_to_file("final.apkg")
```
