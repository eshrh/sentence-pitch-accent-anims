```python
pitchdb = pd.read_csv("accents.txt",sep="\t",names=["kanji","kana","pitch"])
```

```python
def format_pitch(word,pit):
    s = "<span>"
    for n,i in enumerate(word):
        if n+1 in pit:
            s +='<span style="display:inline-block;position:relative;padding-right:0.1em;margin-right:0.1em;">'\
                 + i + \
                 '<span style="display:block;user-select:none;pointer-events:none;position:absolute;top:0.1em;left:0;right:0;height:0;border-top:0.1em solid;right:-0.1em;height:0.4em;border-right:0.1em solid;"></span>' \
                 + "</span></span>"
        else:
            s +='<span style="display:inline-block;position:relative;">' + i + '</span>'
    return s+"</span>"
```

```python
pithtml = []
for i,sent in enumerate(sents):
    words = wakati.parse(sent).split()
    senthtml = ""
    for word in words:
        pitches = pitchdb[pitchdb["kanji"]==word]["pitch"].values
        if pitches.size==0:
            senthtml+=word
            continue
        pitches = pitches[0]
        pitches = [int(re.sub(r'\([^)]*\)', '',i)) for i in pitches.split(",")]
        senthtml+=format_pitch(word,pitches)
    pithtml.append(senthtml)
    print(i,2000)
```
